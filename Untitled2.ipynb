```python
!pip install diffusers transformers accelerate torch safetensors --quiet

import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# SAFETY CHECKER DISABLED (to avoid false NSFW blocks)
# ------------------------------------------------------------
def dummy_checker(images, **kwargs):
    return images, [False] * len(images)

# ------------------------------------------------------------
# LOAD BASE MODEL
# ------------------------------------------------------------
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    safety_checker=None
).to("cuda")

pipe.safety_checker = dummy_checker

# ------------------------------------------------------------
# FUNCTION TO FORCE MEDICAL-STYLE OUTPUT
# ------------------------------------------------------------
def to_medical_prompt(user_prompt):

    # Medical imaging template
    medical_template = f"""
    highly detailed medical imaging scan,
    clinical style, diagnostic quality,
    grayscale or blue-tone,
    realistic anatomy,
    {user_prompt},
    shown as one of: X-ray, MRI, CT, Ultrasound, Microscopy or Pathology slide.
    """

    return medical_template

# ------------------------------------------------------------
# NEGATIVE PROMPT TO BLOCK NON-MEDICAL OUTPUT
# ------------------------------------------------------------
medical_negative = """
cartoon, painting, digital art, human photography, animals,
fantasy, scenery, colorful backgrounds, unrealistic anatomy,
3d render, anime, illustration, fashion, people posing
"""

# ------------------------------------------------------------
# USER PROMPT → MEDICAL PROMPT
# ------------------------------------------------------------
user_prompt = "MRI brain scan showing a coronal view of a glioma"  # <<< change this anytime
final_prompt = to_medical_prompt(user_prompt)

# ------------------------------------------------------------
# GENERATE IMAGE
# ------------------------------------------------------------
image = pipe(
    final_prompt,
    negative_prompt=medical_negative,
    num_inference_steps=35,
    guidance_scale=8
).images[0]

# ------------------------------------------------------------
# DISPLAY
# ------------------------------------------------------------
plt.figure(figsize=(6,6))
plt.imshow(image)
plt.axis("off")
plt.show()

# SAVE
image.save("medical_image.png")
print("✔ Saved as medical_image.png")
```

```python
!pip install -q diffusers[torch] transformers accelerate safetensors einops torchvision ftfy gdown

import os
import math
import numpy as np
from PIL import Image, ImageFilter, ImageOps
import matplotlib.pyplot as plt
import torch

# -------------------------
# CONFIG: change if needed
# -------------------------
HF_MODEL_ID = "runwayml/stable-diffusion-v1-5"   # general SD base (works for medical prompts)
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
USE_SAFETY_CHECKER = False      # we intentionally disable to avoid false blocks
LOCAL_LORA_PATH = None          # set to "/content/my_medical_lora.safetensors" if you have it locally
FALLBACK_SYNTHETIC = True       # use synthetic generator if model load fails

# -------------------------
# Helper: disable safety
# -------------------------
def dummy_checker(images, **kwargs):
    return images, [False] * len(images)

# -------------------------
# Try loading the SD pipeline
# -------------------------
sd_pipe = None
try:
    from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler
    # load model
    print("Loading Stable Diffusion pipeline (this may take a minute)...")
    sd_pipe = StableDiffusionPipeline.from_pretrained(
        HF_MODEL_ID,
        torch_dtype=torch.float16 if DEVICE=="cuda" else torch.float32,
        safety_checker=None if not USE_SAFETY_CHECKER else None
    ).to(DEVICE)
    sd_pipe.scheduler = DPMSolverMultistepScheduler.from_config(sd_pipe.scheduler.config)
    if not USE_SAFETY_CHECKER:
        sd_pipe.safety_checker = dummy_checker
    print("-> SD pipeline loaded on", DEVICE)
except Exception as e:
    print("Warning: could not load SD pipeline:", str(e))
    sd_pipe = None

# -------------------------
# Optional: LoRA loader (apply via text encoder + unet alpha)
# NOTE: requires safetensors LoRA in diffusers format or adapted loader.
# We'll implement a simple merge-from-local checkpoint if provided (lightweight)
# -------------------------
def apply_lora(pipe, lora_path, alpha=0.8):
    """
    Simple LoRA merging: this requires the LoRA to be provided in diffusers-compatible state-dict.
    If not available, skip. This is a placeholder: users who have LoRA files in known format can
    adapt this part.
    """
    if not lora_path or not os.path.exists(lora_path):
        print("No LoRA applied (no file).")
        return pipe
    try:
        from safetensors.torch import load_file
        lora_sd = load_file(lora_path)
        # naive merge: add scaled weights to UNet (best-effort; many LoRAs require special handling)
        unet = pipe.unet
        unet_sd = unet.state_dict()
        merged = False
        for k,v in lora_sd.items():
            if k in unet_sd:
                unet_sd[k] = unet_sd[k] + alpha * v.to(unet_sd[k].device)
                merged = True
        if merged:
            unet.load_state_dict(unet_sd)
            print("LoRA merged into UNet (alpha=%s)" % alpha)
        else:
            print("LoRA file did not match UNet keys; skipping merge.")
    except Exception as e:
        print("LoRA apply failed:", e)
    return pipe

if sd_pipe and LOCAL_LORA_PATH:
    sd_pipe = apply_lora(sd_pipe, LOCAL_LORA_PATH, alpha=0.8)

# -------------------------
# Strong medical prompt template & negative prompt
# -------------------------
def medical_template(user_prompt, modality="auto", style="realistic", view=None):
    """
    Build a medical-focused prompt that forces the model into medical imaging domain.
    modality: 'ct','mri','xray','ultrasound','pathology','endoscopy','ecg','microscopy' or 'auto'
    view: optional additional text like 'axial','sagittal','coronal','pa view'
    """
    base = "high-resolution clinical medical imaging, diagnostic quality, photorealistic scan, anatomical accuracy, "
    if modality != "auto":
        base += f"{modality} imaging, "
    else:
        base += "medical imaging (CT, MRI, X-ray, ultrasound, pathology, microscopy), "
    if view:
        base += f"{view}, "
    base += "grayscale or clinically appropriate coloring, realistic medical imaging artifacts (gantry markers, scale bars, noise, windowing), "
    base += user_prompt
    # enforce that result should be shown as a clinical image
    base += ". Render as clinical diagnostic image, no artistic styles or fantasy."
    return base

MEDICAL_NEGATIVE = (
    "cartoon, painting, illustration, anime, fantasy, surreal, colorful background, fashion, portrait photography, people posing, studio lighting"
)

# -------------------------
# Post-processing per modality
# -------------------------
def postprocess_image_pil(img_pil, modality):
    """
    img_pil: PIL.Image RGB
    returns processed PIL.Image in medical style for modality
    """
    img = img_pil.convert("RGB")
    w,h = img.size

    if modality in ("ct","ct-scan"):
        # Convert to grayscale, apply windowing-like contrast and add subtle noise
        img = ImageOps.grayscale(img)
        arr = np.array(img).astype(np.float32)
        # apply CT-like window center/width (example: soft tissue)
        center, width = 40.0, 300.0
        arr = (arr - (center + 128)) * (255.0/width) + 128
        arr = np.clip(arr, 0, 255).astype(np.uint8)
        img = Image.fromarray(arr)
        img = img.filter(ImageFilter.UnsharpMask(radius=1, percent=80, threshold=3))
        # add faint film grain
        noise = np.random.normal(0, 6, (h,w)).astype(np.uint8)
        arr = np.array(img)
        arr = np.clip(arr + noise, 0, 255).astype(np.uint8)
        return Image.fromarray(arr).convert("L").convert("RGB")

    if modality in ("mri","mri-scan"):
        img = ImageOps.grayscale(img)
        arr = np.array(img).astype(np.float32)
        # increase mid-range contrast for T2-like bright fluid
        arr = (arr - arr.mean()) * 1.2 + arr.mean()
        arr = np.clip(arr, 0, 255).astype(np.uint8)
        img = Image.fromarray(arr)
        img = img.filter(ImageFilter.GaussianBlur(radius=0.8))
        # add ring artifact vignette
        mask = Image.new("L", img.size, 0)
        draw = Image.new("L", img.size, 0)
        # vignette created by radial gradient
        xv, yv = np.meshgrid(np.linspace(-1,1,w), np.linspace(-1,1,h))
        radial = np.sqrt(xv**2 + yv**2)
        vignette = (1.0 - np.clip(radial, 0, 1))**1.5
        arr = np.array(img).astype(np.float32)
        arr = arr * (0.8 + 0.4*vignette)
        arr = np.clip(arr, 0, 255).astype(np.uint8)
        return Image.fromarray(arr).convert("L").convert("RGB")

    if modality in ("xray","x-ray","xr"):
        # high-contrast grayscale with "film" look
        img = ImageOps.grayscale(img).resize((w,h))
        arr = np.array(img).astype(np.float32)
        arr = 255 - arr  # invert to mimic bone bright on white background sometimes
        arr = np.clip((arr - arr.mean()) * 1.4 + arr.mean(), 0, 255).astype(np.uint8)
        img = Image.fromarray(arr)
        # add film scratches / ribs simulation lightly (optional)
        img = img.filter(ImageFilter.UnsharpMask(radius=1.2, percent=120, threshold=2))
        return img.convert("RGB")

    if modality in ("ultrasound","us","sonography"):
        # convert to grayscale, add speckle and typical ultrasound blur
        img = ImageOps.grayscale(img)
        arr = np.array(img).astype(np.float32)
        # speckle: multiplicative noise
        speckle = 1 + (np.random.randn(*arr.shape) * 0.08)
        arr = arr * speckle
        arr = cvt = np.clip(arr, 0, 255).astype(np.uint8)
        img = Image.fromarray(cvt).filter(ImageFilter.GaussianBlur(radius=1.2))
        # add bottom fade to mimic probe shadow
        fade = np.linspace(1.0, 0.6, h).reshape(h,1)
        arr = np.array(img).astype(np.float32) * fade
        arr = np.clip(arr, 0, 255).astype(np.uint8)
        return Image.fromarray(arr).convert("RGB")

    if modality in ("pathology","histology","h&e","histopath"):
        # attempt H&E stain colorization: map grayscale to H/E colors
        img = ImageOps.grayscale(img).resize((w,h))
        arr = np.array(img).astype(np.float32)/255.0
        # simple fake H&E: use two color channels mixing
        he = np.zeros((h,w,3), dtype=np.float32)
        he[:,:,0] = arr * 0.6 + 0.4*arr**1.1   # hematoxylin-ish (blue-purple)
        he[:,:,1] = arr * 0.4                  # eosin-ish (pink)
        he[:,:,2] = arr * 0.85
        he = np.clip(he*255,0,255).astype(np.uint8)
        return Image.fromarray(he)

    # default: return original resized to medical-friendly look (grayscale + vignette)
    img = ImageOps.grayscale(img)
    arr = np.array(img).astype(np.float32)
    arr = np.clip((arr - arr.mean()) * 1.1 + arr.mean(), 0, 255).astype(np.uint8)
    return Image.fromarray(arr).convert("RGB")

# -------------------------
# Procedural fallback generator (higher-quality than earlier)
# -------------------------
def procedural_medical_generator(prompt, modality="auto", seed=None, size=512):
    """
    Generates a plausible medical-looking image from prompt keywords without external models.
    Used as fallback when SD is not available.
    """
    if seed is not None:
        np.random.seed(seed)
    canvas = np.random.normal(128, 18, (size, size)).astype(np.uint8)
    img = Image.fromarray(canvas).convert("L")
    draw = Image.fromarray(canvas.copy()).convert("L")
    # create central anatomy circle/shape
    center = (size//2, size//2)
    import cv2
    arr = np.array(img)
    cv2.circle(arr, center, int(size*0.38), (160), -1)
    # modality-specific fake lesions / nodules
    p = prompt.lower()
    if "tumor" in p or "lesion" in p:
        cv2.circle(arr, (center[0]-int(size*0.08), center[1]+int(size*0.05)), int(size*0.07), (230), -1)
    if "nodule" in p:
        cv2.circle(arr, (center[0]+int(size*0.15), center[1]-int(size*0.05)), int(size*0.04), (200), -1)
    img = Image.fromarray(arr).filter(ImageFilter.GaussianBlur(radius=1.5))
    # apply modality postprocessing
    return postprocess_image_pil(img.convert("RGB"), modality if modality!="auto" else "mri")

# -------------------------
# Main generator function
# -------------------------
def generate_medical_from_text(user_prompt,
                               modality="auto",   # 'mri','ct','xray','ultrasound','pathology','endoscopy','microscopy','ecg','auto'
                               view=None,
                               guidance_scale=7.5,
                               steps=30,
                               seed=None,
                               use_lora=False,
                               lora_path=None):
    """
    Returns PIL.Image
    """
    if seed is not None:
        torch.manual_seed(seed)
        np.random.seed(seed)

    # Build forced medical prompt
    final_prompt = medical_template(user_prompt, modality=modality, view=view)
    print("Final medical prompt:\n", final_prompt)

    # If pipeline is available, try to generate
    if sd_pipe is not None:
        try:
            # optionally apply LoRA if specified
            if use_lora and lora_path:
                apply_lora(sd_pipe, lora_path, alpha=0.8)

            result = sd_pipe(
                final_prompt,
                negative_prompt=MEDICAL_NEGATIVE,
                num_inference_steps=steps,
                guidance_scale=guidance_scale
            )
            img = result.images[0]
            # Postprocess to enforce modality style
            img = postprocess_image_pil(img, modality if modality!="auto" else detect_modality_from_prompt(user_prompt))
            return img
        except Exception as e:
            print("SD generation failed, falling back to procedural generator. Error:", e)

    # Fallback: procedural
    print("Using fallback procedural generator.")
    return procedural_medical_generator(user_prompt, modality if modality!="auto" else detect_modality_from_prompt(user_prompt), seed=seed)

# Simple heuristic to detect modality from prompt
def detect_modality_from_prompt(prompt):
    p = prompt.lower()
    if any(k in p for k in ["mri", "mri-scan", "t1", "t2", "flair"]):
        return "mri"
    if any(k in p for k in ["ct","ct-scan","computed tomography"]):
        return "ct"
    if any(k in p for k in ["x-ray","xray","chest xray","radiograph"]):
        return "xray"
    if any(k in p for k in ["ultrasound","us","sonography","echo"]):
        return "ultrasound"
    if any(k in p for k in ["histology","histopathology","h&e","pathology","slide"]):
        return "pathology"
    if any(k in p for k in ["endoscopy","gastroscopy","colonoscopy","laparoscopy"]):
        return "endoscopy"
    return "mri"

# -------------------------
# USAGE: change these variables below
# -------------------------
user_prompt = "cat"
modality = "auto"      # set 'auto' to detect, or explicitly 'mri','ct','xray','ultrasound','pathology','endoscopy'
view = "axial"
seed = 42
use_lora = False
lora_path = LOCAL_LORA_PATH  # if you have a local LoRA file, set path here

# -------------------------
# Generate
# -------------------------
img = generate_medical_from_text(user_prompt, modality=modality, view=view, seed=seed,
                                guidance_scale=8.0, steps=30, use_lora=use_lora, lora_path=lora_path)

# show and save
plt.figure(figsize=(6,6))
plt.imshow(img.convert("L"), cmap="gray")
plt.axis("off")
plt.title(f"{modality.upper()} - {user_prompt}")
plt.show()

out_path = "/content/advanced_medical_output.png"
img.save(out_path)
print("Saved to", out_path)

```
